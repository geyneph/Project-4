```python
#None Gene influence data frame
noneGeneDF = df_mutation[[
 'age_at_diagnosis',
 'chemotherapy',
 'hormone_therapy',
 'mutation_count',
 'tumor_size',
 'tumor_stage',
 'overall_survival',
 ]]
```

```python
noneGeneDF = noneGeneDF.dropna()
noneGeneDF
```

```python
# Separate the data into labels and features
y = noneGeneDF[['overall_survival']]
X = noneGeneDF.copy()
X.drop('overall_survival',inplace=True,axis=1)
X.drop('patient_id',inplace=True,axis=1)
```

```python
#display y
y.head()
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
model = LogisticRegression(random_state=1)
# Fit the model using training data
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
list(predictions)
```

```python
cmatrix = confusion_matrix(y_test, predictions)
cmatrix_df = pd.DataFrame(cmatrix)  
cmatrix_df
```

```python
# Print the classification report for the model
print(f'Model Score: {round(model.score(X_test, y_test),2)}')
print('-'*100)
print(classification_report(y_test, predictions))
print('-'*100)
```

```python
X_test['real'] = y_test
X_test['prediction']=predictions
X_test.head(50)
```

```python
#Now We train model Again 
y = result['overall_survival']
X = result.copy()
X.drop('overall_survival',inplace=True,axis=1)
```

```python
predictions = model.predict(X_test)
proba = model.predict_proba(X_test)
proba = proba.tolist()
proba
```
